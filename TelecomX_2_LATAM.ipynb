```python
# Estandarizaci贸n de los encabezados de las columnas
import re

# Funci贸n para estandarizar los nombres de las columnas
def estandarizar_columna(nombre):
    nombre = re.sub(r'(?<=[a-z])(?=[A-Z])', '_', nombre)  # separa camel case, Reemplazando cada coincidencia del patr贸n con un guion bajo (_) en la variable nombre.
    nombre = nombre.replace('.', '_')                     # reemplaza puntos
    nombre = nombre.strip().lower()                       # min煤sculas y sin espacios
    return nombre

# Aplicar la funci贸n a los nombres de las columnas
df.columns = [estandarizar_columna(col) for col in df.columns]
df.columns
```

```python
df.head()
```

```python
# Estandarizaci贸n en los registros de las columnas 'object'

for col in df.select_dtypes(include='object').columns:
    df[col] = (
        df[col]
        .astype(str)                                 # Asegura strings
        .str.strip()                                 # Elimina espacios al inicio/final
        .str.lower()                                 # Convierte a min煤sculas
        .str.replace(r'[\s\-]+', '_', regex=True)    # Reemplaza espacios y guiones medio por guion bajo
    )

df
```
